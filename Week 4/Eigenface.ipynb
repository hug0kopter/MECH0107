```python
#import necessary libraries
import numpy as np
import scipy as scipy
import matplotlib.pyplot as plt
import imageio.v2 as imageio
from skimage import transform, color
import glob
import os
```

```python
# Step 1: Get all image file paths
image_paths = glob.glob("faces/*.jpg")

# Step 2: Load and resize images using a for loop
images = {}  # Dictionary to store images

for path in image_paths:
    # Read the image
    img = imageio.imread(path)
    
    # Resize the image to 120x80
    img_resized = transform.resize(img, (120, 80), anti_aliasing=True)
    img_doubled = np.asarray(img_resized)
    img_grayscaled = color.rgb2gray(img_doubled)
    
    # Store in dictionary with filename as key
    images[path] = img_grayscaled

# Step 3: Display the first image as a check
first_image_key = image_paths[0]  # Get the first image path
plt.figure()
plt.imshow(images[first_image_key], cmap='gray')
plt.axis('off')
plt.show()

# Step 4: Print confirmation message
print(f"Loaded {len(images)} images, all resized to {images[first_image_key].shape}")
```

```python
# Step 5: Filter only Clooney images
clooney_images = []  # List to store Clooney images

for path, img in images.items():
    filename = os.path.basename(path)  # Extract only filename
    if "clooney" in filename:
        clooney_images.append(img)  # Add to list

# Step 6: Convert list to NumPy array
clooney_stack = np.array(clooney_images) 

# Step 7: Compute the mean across all Clooney images
average_clooney = np.mean(clooney_stack, axis=0)

# Step 8: Display the averaged Clooney face
plt.figure()
plt.imshow(average_clooney, cmap='gray')
plt.axis('off')
plt.title("Average Clooney Face")
plt.show()
```

```python
# Define the names we want to process
names = ["clooney", "thatcher", "obama", "fed"]

# Dictionary to store average images
average_faces = {}

# Loop through each name
for name in names:
    # Step 5: Filter images for the current name
    selected_images = []  # List to store images for the current person

    for path, img in images.items():
        filename = os.path.basename(path)  # Extract only filename
        if name in filename:
            selected_images.append(img)  # Add to list

    # Step 6: Convert list to NumPy array
    if len(selected_images) > 0:  # Avoid division by zero
        image_stack = np.array(selected_images)  # Shape: (num_images, 120, 80)

        # Step 7: Compute the mean across all images
        average_face = np.mean(image_stack, axis=0)  # Shape: (120, 80)

        # Store in dictionary
        average_faces[name] = average_face
        
#show the image
plt.figure

plt.subplot(2,2,1)
plt.imshow(average_faces["clooney"], cmap='gray')
plt.title("Clooney")

plt.subplot(2,2,2)
plt.imshow(average_faces["thatcher"], cmap='gray')
plt.title("thatcher")

plt.subplot(2,2,3)
plt.imshow(average_faces["obama"], cmap='gray')
plt.title("obama")

plt.subplot(2,2,4)
plt.imshow(average_faces["fed"], cmap='gray')
plt.title("fed")
```

```python
# Step 1: Create a dictionary to store the D matrices for each person
D_matrices = {}

# Step 2: Reshape all images into column vectors and stack them into row matrices
m = 120  # Height of images
n = 80   # Width of images

for name in names:
    vectorised_images = []  # List to store all row vectors for this person

    for path, img in images.items():
        filename = os.path.basename(path)  # Extract only filename
        if name in filename:
            # Reshape each image into a single row vector
            vectorised_face = np.reshape(img, (1, m * n))
            vectorised_images.append(vectorised_face)  # Append to list

    # Convert list to NumPy array (stacking all images into one matrix)
    if len(vectorised_images) > 0:  # Ensure we have images before stacking
        D_matrices[name] = np.vstack(vectorised_images)  # Stack all images for this person

        # Print shape to verify
        print(f"{name} D matrix shape:", D_matrices[name].shape)

# Example: Accessing the D matrix for Clooney
print("Clooney D matrix:", D_matrices["clooney"].shape if "clooney" in D_matrices else "No images found")

```

```python
covariance_dictionary = {}

for name in names:
    D_transposed = np.transpose(D_matrices[name])
    D = D_matrices[name]
    covariant_matrix = np.matmul(D_transposed, D)
    covariance_dictionary[name] = covariant_matrix
    print(f'Covariance matrix size for {name} is: {covariant_matrix.shape}')
```

```python
eigvals_dict = {}
eigvecs_dict = {}

for name in names:
    print(f'{name}face')
    eigvals, eigvecs = scipy.sparse.linalg.eigs(covariance_dictionary[name], 20, which='LM')
    eigvals_dict[name] = eigvals
    eigvecs_dict[name] = eigvecs
```
