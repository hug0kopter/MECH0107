```python
import numpy as np
import matplotlib.pyplot as plt
from pyDOE import lhs  # Use pyDOE2 instead of pyDOE

# Define parameters
n_samples = 10  # Number of samples

# Generate LHS samples using pyDOE2
lhs_samples = lhs(2, samples=n_samples, criterion='maximin')

# Scale feature 1 to [4, 10]
feature_1_samples = lhs_samples[:, 0] * 6 + 4  

# Scale feature 2 to [0, 5]
feature_2_samples = lhs_samples[:, 1] * 5  

# Combine features into a NumPy array
new_lhs_samples = np.column_stack((feature_1_samples, feature_2_samples))

# Plot the samples
plt.scatter(new_lhs_samples[:, 0], new_lhs_samples[:, 1], color='blue', label="LHS Samples")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.xlim(4, 10)  # Updated limits to match scaling
plt.ylim(0, 5)

plt.legend()
plt.show()

# Print generated samples
print("LHS Samples:\n", new_lhs_samples)
```
